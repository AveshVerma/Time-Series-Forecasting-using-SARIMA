### Volume Traded and Interpretation

Volume represents the number of stocks, futures or options contracts, which are traded during a certain period of time, most often a day. The higher the volume, the more active the instrument we are trading is. Every unit of volume in any market reflects the actions taken by two sides: one trader buys a given share or contract and another trader sells a given share or contract

Now we will analyse the total volume traded by each company.

```python
tata['Volume'].plot(label = 'TATA', figsize = (12, 6))
msuzuki['Volume'].plot(label = 'MARUTI SUZUKI', figsize = (12, 6))
bajajauto['Volume'].plot(label = 'BAJAJ AUTO', figsize = (12, 6))

plt.title('Volume Trade of TATA, Maruti Suzuki & Bajaj Auto')
plt.ylabel('Volume Traded')
plt.legend(['TATA', 'MARUTI SUZUKI', 'BAJAJ AUTO'])
plt.show()
```

### Market Capitalization

Market capitalization refers to the total dollar market value of a company's outstanding shares of stock. Commonly referred to market cap, it is calculated by multiplying the total number of a company's outstanding shares by the current market price of one share.

```python
tata['Total Trade'] = tata['Open'] * tata['Volume']
msuzuki['Total Trade'] = msuzuki['Open'] * msuzuki['Volume']
bajajauto['Total Trade'] = bajajauto['Open'] * bajajauto['Volume']
```

```python
tata['Total Trade'].plot(label = 'TATA', figsize = (14,7))
msuzuki['Total Trade'].plot(label='MARUTI SUZUKI')
bajajauto['Total Trade'].plot(label='BAJAJ AUTO')
plt.title('Total Money Traded by TATA, MARUTI SUZUKI & BAJAJ AUTO')
plt.legend()
plt.ylabel('Total Trade')
plt.show()
```

We see that TATA Motors has been traded maximum amount in the month October-November, 2021 while Maruti Suzuki has been traded maximum around January, 2022. 

```python
msuzuki_trade = msuzuki['Total Trade'].argmax()
msuzuki.iloc[msuzuki_trade]
```

Therefore, on 27th Jan, 2022 - Maruti Suzuki has been traded the maximum since last 1 year.

### Moving Average


Moving average is a technique to get an overall idea of the trends in a data set, an average of any subset of numbers. It is the best practice that Analyzers use when studing certain trends. Moving average can be calculated for specific interval of time, say 20 days, 50 days, 100 days and so on.

```python
tata['MA50'] = tata['Open'].rolling(50).mean()
tata['MA100'] = tata['Open'].rolling(100).mean()
tata['MA150'] = tata['Open'].rolling(150).mean()

msuzuki['MA50'] = msuzuki['Open'].rolling(50).mean()
msuzuki['MA100'] = msuzuki['Open'].rolling(100).mean()
msuzuki['MA150'] = msuzuki['Open'].rolling(150).mean()
```

```python
tata['Open'].plot(label='Open Price',figsize = (15,6))
tata['MA50'].plot(label='MA50')
tata['MA100'].plot(label='MA100')
tata['MA150'].plot(label='MA150')
plt.legend()
```

### Daily percentage change


Percentage change is a mathematical concept that represents the degree of change over time. It is used for many purposes in finance, often to represent the price change of a security.

This formula returns the % gain or % loss

Percentage Change = (Today's Closing price/ Yesterday's closing price)-1

```python
tata['percent_change'] = (tata['Close']/tata['Close'].shift(1)) -1 
msuzuki['percent_change'] = (msuzuki['Close']/msuzuki['Close'].shift(1)) -1 
bajajauto['percent_change'] = (bajajauto['Close']/bajajauto['Close'].shift(1)) -1 
```

```python
from pandas.plotting import scatter_matrix

tata['percentage_change'] = (tata['Close']/tata['Close'].shift(1)) -1 
tata_change = tata[['percentage_change']]
scatter_matrix(tata_change,hist_kwds={'bins':40})
```

### Highest Returns

```python
companies_return = pd.concat([tata['percent_change'],msuzuki['percent_change'],bajajauto['percent_change']], axis = 1)

companies_return.columns = ['TATA Return','Maruti Suzuki Return','Bajaj Auto Return']

companies_return.plot(figsize = (14,7))
```

Here we can see that TATA has given highest return of 20% in the month of Oct,21 while Maruti has been giving consistent returns of 5-6% after every 2 months. Bajaj Auto on the other hand remains a low-return company.

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0,1))
df1 = sc.fit_transform(np.array(df1).reshape(-1,1))
```

```python
##splitting dataset into train and test split
training_size=int(len(df1)*0.65)
test_size=len(df1)-training_size
train_data,test_data=df1[0:training_size,:],df1[training_size:len(df1),:1]
```

```python
import numpy

def create_dataset(dataset, time_step = 1):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_step-1):
        a = dataset[i:i+time_step, 0]
        dataX.append(a)
        dataY.append(dataset[i+time_step, 0])
    return numpy.array(dataX), numpy.array(dataY)
```

```python
# reshape into X=t,t+1,t+2,t+3 and Y=t+4
time_step = 100
X_train, y_train = create_dataset(train_data, time_step)
X_test, ytest = create_dataset(test_data, time_step)
```
